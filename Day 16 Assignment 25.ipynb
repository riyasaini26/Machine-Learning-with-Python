```python
import pandas as pd
import os
```

```python
os.chdir('D:\\RIYA\\college stuff\\ML with python')
print (os.getcwd())

 
df = pd.read_csv("College.csv")
print (df)

```

```python
print (df.shape)
print (df.columns)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X, y = df.iloc[:, 1:].values, df.iloc[:, 0].values

# Encode label category  male -> 1 and  female -> 0

target_encoder = LabelEncoder()
y = target_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
```

```python
from sklearn.svm import LinearSVC,SVC
classifier = LinearSVC()


classifier.fit(X_train,y_train)

y_predict = classifier.predict(X_test)

print (classifier.score(X_test,y_test))

```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_predict,y_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))

```

```python
classifier = SVC(kernel='linear')
classifier.fit(X_train,y_train)
classifier.score(X_test,y_test)
```

```python
y_predict = classifier.predict(X_test)
print(confusion_matrix(y_predict,y_test))
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
print(classification_report(y_test, y_predict))

```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler

```

```python
scaler = StandardScaler()
X, y = df.iloc[:, 1:].values, df.iloc[:, 0].values
X = scaler.fit_transform(X)
```

```python
target_encoder = LabelEncoder()
y = target_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
```

```python
classifier = SVC()
classifier.fit(X_train,y_train)
classifier.score(X_test,y_test)
```

```python
y_predict = classifier.predict(X_test)
print(confusion_matrix(y_predict,y_test))
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
print(classification_report(y_test, y_predict))
```

```python
import numpy as np

C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
param_grid = dict( gamma=gamma_range,C=C_range)
grid = GridSearchCV(SVC(), param_grid=param_grid,verbose=2)
grid.fit(X_train, y_train)
```

```python
print("The best parameters are %s with a score of %0.2f"
      % (grid.best_params_, grid.best_score_))
```

```python

y_predict = grid.predict(X_test)
print(confusion_matrix(y_predict,y_test))
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
print(classification_report(y_test, y_predict))
```

```python
params = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]}
print (params)
rgrid =  RandomizedSearchCV(SVC(), param_distributions=params, n_iter= 12,cv = 10,  verbose=2 )

rgrid.fit(X_train, y_train)
```

```python
print("The best parameters are %s with a score of %0.2f"
      % (rgrid.best_params_, rgrid.best_score_))
```

```python
y_predict = rgrid.predict(X_test)
print(confusion_matrix(y_predict,y_test))
accuracy = accuracy_score(y_predict,y_test)
print (accuracy )
print(classification_report(y_test, y_predict))
```
