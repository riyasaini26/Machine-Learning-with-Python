```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
math_data['Ethinicity']= label_encoder.fit_transform(math_data['Ethinicity'])
print(math_data.head())

# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
ds_data['Ethinicity']= label_encoder.fit_transform(ds_data['Ethinicity'])
print(ds_data.head())

# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
phy_data['Ethinicity']= label_encoder.fit_transform(phy_data['Ethinicity'])
print(phy_data.head())
```

```python
ds_data = ds_data.drop(['Name'], axis=1)
print(ds_data.head())

```

```python

phy_data = phy_data.drop(['Name'], axis=1)
print(phy_data.head())
```

```python
# Fill the missing score for a student 
# to the average of the class
print (math_data.isnull().sum())
mean1 = math_data["Score"].mean()
math_data['Score']=math_data['Score'].fillna(mean1)
print (math_data.isnull().sum())
```

```python
print (ds_data.isnull().sum())
mean1 = ds_data["Score"].mean()
ds_data['Score']=ds_data['Score'].fillna(mean1)
print (ds_data.isnull().sum())
```

```python
print (phy_data.isnull().sum())
mean1 = phy_data["Score"].mean()
phy_data['Score']=phy_data['Score'].fillna(mean1)
print (phy_data.isnull().sum())
```

```python
all_data =[math_data,ds_data,phy_data]
all_data
```

```python
# Convert Sex M to 1 and F to 2

for dataset in all_data:
    dataset['Sex'] = dataset['Sex'].map({'M': 1, 'F': 2}).astype(int)


all_data_df = pd.concat(all_data)
all_data_df
```
