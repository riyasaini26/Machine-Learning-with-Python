```python
allcategory_cols = dtype_null[dtype_null.Datatype=='object']
allcategory_cols = list(allcategory_cols.index)
allcategory_cols
```

```python
df1=df
for i in allcategory_cols:
    df1 = pd.get_dummies(df1,columns=[i],drop_first=True)
df1.to_excel('D:\\RIYA\\college stuff\\ML with python\\CS32onehot.xlsx')
df1
```

```python
target.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
target = label.fit_transform(target)
print(target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1, target, test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

```

```python
ydt_pred = dt.predict(X_test)
ydt_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ydt_pred,y_test)
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)  
```

```python
yrf_predict = rf.predict(X_test)
yrf_predict
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(yrf_predict,y_test)
print(accuracy)
```
